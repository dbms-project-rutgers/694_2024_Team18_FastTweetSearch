```python
print(type(df))
df1 = df.fillna('MISSING_INFORMATION')
df1 = df1.replace('','MISSING_INFORMATION').rename(columns = {'profile_image_url\t':'profile_image_url'})
df1['created_at'] = pd.to_datetime(df1['created_at']).fillna(pd.Timestamp('2099-12-31 09:13:33+00:00'))
df1['created_at'] = df1['created_at'].astype(str)
# df = df.select_dtypes(include='object').fillna('unknown')
cols = [i for i in df1.columns]
cols[:-1]
```

```python
import json
import psycopg2

# PostgreSQL connection details
PGHOST = "twitter.postgres.database.azure.com"
PGUSER = "neeraj"
PGPORT = 5432
PGDATABASE = "postgres"
PGPASSWORD = "Dbms@123"

# Connect to PostgreSQL
conn = psycopg2.connect(
    dbname=PGDATABASE,
    user=PGUSER,
    password=PGPASSWORD,
    host=PGHOST,
    port=PGPORT
)

# Open a cursor to perform database operations
cur = conn.cursor()

# Create users table if not exists
cur.execute("""
    CREATE TABLE IF NOT EXISTS users (
        id BIGINT PRIMARY KEY,
        name VARCHAR(255),
        screen_name VARCHAR(255),
        location VARCHAR(255),
        url VARCHAR(255),
        followers_count INTEGER,
        friends_count INTEGER,
        statuses_count INTEGER,
        verified BOOLEAN,
        created_at TIMESTAMP,
        profile_background_image_url TEXT,
        profile_background_image_url_https TEXT
    )
""")

# Read JSON string from the variable 'contents' and parse it into a list of dictionaries
users_data = json.loads(contents)

print(users_data[1000])
```

```python
# Define the insert query with ON CONFLICT DO NOTHING
insert_query = """
    INSERT INTO users (
        id, name, screen_name, location, url, followers_count, 
        friends_count, statuses_count, verified, created_at, 
        profile_background_image_url, profile_background_image_url_https
    ) VALUES (
        %(id)s, %(name)s, %(screen_name)s, %(location)s, %(url)s, 
        %(followers_count)s, %(friends_count)s, %(statuses_count)s, 
        %(verified)s, %(created_at)s, %(profile_background_image_url)s, 
        %(profile_background_image_url_https)s
    )
    ON CONFLICT DO NOTHING
"""

# Insert each user data into the PostgreSQL table
for user in users_data:
    try:
        cur.execute(insert_query, user)
    except Exception as e:
        print(f"Error inserting user {user['id']}: {e}")

# Commit the transaction
conn.commit()

# Close communication with the PostgreSQL database
cur.close()
conn.close()
```
